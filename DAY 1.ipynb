```python
Q1 = int_df_mean.quantile(0.25)
Q3 = int_df_mean.quantile(0.75)

IQR = Q3 - Q1

threshold = 1.5

int_df_mean = int_df_mean[~((int_df_mean < (Q1 - threshold * IQR)) | (int_df_mean > (Q3 + threshold * IQR))).any(axis=1)]
print(int_df_mean)
int_df_mean.isnull().sum()
```
