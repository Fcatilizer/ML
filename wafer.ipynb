```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
num_df = df.select_dtypes(include=['int64','float64']).copy()
num_df.head()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
obj_df = obj_df.apply(lb_make.fit_transform)
```

```python
new_df = pd.concat([obj_df, num_df], axis=1)
new_df.head()
```

```python
#fill na with mean
mean_new_df = new_df.fillna(new_df.mean())
```

```python
mean_new_df.isna().sum()
```

```python
mean_new_df.columns
```

```python
#forward feature selection
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression

X = mean_new_df.drop('Good/Bad', axis=1)
y = mean_new_df['Good/Bad']
```

```python
X.head()
```

```python
y.head()
```

```python
print(X.shape), print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape), print(X_test.shape), print(y_train.shape), print(y_test.shape)    
```

```python
sfs1 = sfs(LinearRegression(), k_features=5, forward=True, scoring='r2', cv=5)
sfs1 = sfs1.fit(X_train, y_train)
```
